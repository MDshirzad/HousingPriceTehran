```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('housePrice.csv')
df.info()
df
```


<label style="Fon-weight:Bold;color:#fAff" > There is no missing values </label>

```python



numericalDataFrame = pd.DataFrame(df[["Area","Room","Price","Price(USD)"]])
StringDataFrame = pd.DataFrame(df[["Parking","Warehouse","Elevator","Address"]])
df.describe()
 
```

```python
numericalDataFrame
```


<label style="Fon-weight:Bold;color:#fFAf" > Change Area from object to long </label>

```python

df["Area"]=pd.Series(x.replace(",","") for x in df["Area"])
df["Area"]=df["Area"].astype('Int64')
df["Price"]=df["Price"].astype('Int64')
 
```

<label style="Fon-weight:Bold;color:#fAAf" > Calculate Maximum,Minimum,Mean for Rial ,USD , rooms and area</label>

```python


def drawArea():
    plt.plot(df["Area"])
    plt.plot([0,3500],[df["Area"].max(),df["Area"].max()],color="r")
    plt.plot([0,3500],[df["Area"].mean(),df["Area"].mean()],color="k")
    plt.plot([0,3500],[df["Area"].min(),df["Area"].min()],color="g")
```


<label style="Fon-weight:Bold;color:red" > Removing noises from Area</label>

```python

drawArea()
```

<label style="Fon-weight:Bold;color:#Addf" >Because of the noise, Mean and Min are masked</label>

```python
df=df[(df["Area"]<220) &(df["Area"]>35) ]
df=df[(df["Price"]>100000000) ]
```

```python

drawArea()
```

```python
def drawPrice():
    plt.plot(df["Price"])
    plt.plot([0,3500],[df["Price"].max(),df["Price"].max()],color="r")
    plt.plot([0,3500],[df["Price"].mean(),df["Price"].mean()],color="k")
    plt.plot([0,3500],[df["Price"].min(),df["Price"].min()],color="g")
```

```python
drawPrice()
```
